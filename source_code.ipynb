```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
```

```python
model = SentenceTransformer('stsb-roberta-large')
```

```python
# Actual Evaluation using Transformers
def score(sentence1,sentence2):
    embedding1 = model.encode(sentence1, convert_to_tensor=True)
    embedding2 = model.encode(sentence2, convert_to_tensor=True)
    # compute similarity scores of two embeddings
    cosine_scores = util.pytorch_cos_sim(embedding1, embedding2)
    sc = cosine_scores.item()
    if sc < 0.4:
      return 0
    elif sc > 0.4 and sc < 0.65:
      return 0.5
    else:
      return 1
```

```python
# Extracting text from pdf
from PyPDF2 import PdfReader

# creating a pdf reader object
reader1 = PdfReader('Sushmita_15.pdf')
reader2 = PdfReader('Science_Answers.pdf')
# printing number of pages in pdf file 

# getting a specific page from the pdf file
page1 = reader1.pages[0] #getting first page
page2 = reader2.pages[0]

# extracting text from page
text1 = page1.extract_text()
text2 = page2.extract_text()
print(text1)
print(text2)
```

```python
# list to string
def listToString(s):
 
    # initialize an empty string
    str1 = ""
 
    # traverse in the string
    for ele in s:
        str1 += ele+" "
 
    # return string
    return str1
```

```python
que = int(input("Enter the no. of questions"))    #(8) 3,2,2,3,2,1,2,5
total = 0
que_mark = []
for i in range(que):
    mark = int(input("enter the marks for question no. "+str(i+1)+":- "))
    que_mark.append(mark)
    total = total+mark
#print(que_mark)
```

```python
#extracting answers from both the pdf and evaluating them question wise
textls1 = []
textls2 = []
textls1 = text1.split(" ")
textls2 = text2.split(" ")
sum = 0
for i in range(1,que+1):
    print(" ---------------")
    print("| Question no. "+str(i)+"|")
    sent1 = []
    sent2 = []
    mystr = "\n"+"Ans"+str(i)
    if mystr in textls1:
        mystr2 = "\n"+"Ans"+str(i+1)
        ind = textls1.index(mystr)
        for j in range(ind+1,len(textls1)):
            if textls1[j]!=mystr2:
                sent1.append(textls1[j])
            else:
                break
    sentence1 = listToString(sent1)
    print("Student answer"+str(i)+"---> "+sentence1)
    if mystr in textls2:
        mystr2 = "\n"+"Ans"+str(i+1)
        ind = textls2.index(mystr)
        for j in range(ind+1,len(textls2)):
            if textls2[j]!=mystr2:
                sent2.append(textls2[j])
            else:
                break
    sentence2 = listToString(sent2)
    print("Actual Answer"+str(i)+"---> " +sentence2)
    print("------------------------------------------------------")
    mrk = que_mark[i-1]*score(sentence1,sentence2)
    sum = sum+mrk
    print("||Score for this question :-----> "+str(mrk)+"/"+str(que_mark[i-1]))
    print("------------------------------------------------------")

    print("\n\n")
print("\nTotal score:- "+str(sum)+"/"+str(total))

```
